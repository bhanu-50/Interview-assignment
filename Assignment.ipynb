# **Retention & Engagement**

### **For each player, count days active in the first 30 days after registration.**

```
Player_Details.head()
```

```
Player_Activity['ActivityMonth'] = pd.to_datetime(Player_Activity['ActivityMonth'])
```

```
Player_Activity.head()
```

```
player_act_detail = Player_Details.merge(
    Player_Activity,
    left_on='Src_Player_Id',
    right_on='src_player_id',
    how='inner'
)
```

```
player_act_detail.head()
```

```
Active_in_first_30_days = player_act_detail[player_act_detail['ActivePlayerDays']>=30]
Active_in_first_30_days = Active_in_first_30_days.reset_index(drop=True)
```

```
Active_in_first_30_days.head()
```

```
player_act_detail['Signup_Date'] = pd.to_datetime(player_act_detail['Signup_Date'])
player_act_detail['ActivityMonth'] = pd.to_datetime(player_act_detail['ActivityMonth'])

player_act_detail['days_after_signup'] = (player_act_detail['ActivityMonth'] - player_act_detail['Signup_Date']).dt.days

first_30_days = player_act_detail[player_act_detail['days_after_signup'] < 30]

player_active_days_30 = first_30_days.groupby('Src_Player_Id')['ActivePlayerDays'].sum().reset_index()
```

```
player_active_days_30.sort_values(by='ActivePlayerDays', ascending=False).head()
```

**Group players into cohorts**

```
bins = [1, 10, 20, 40, 60, 90]
labels = ['1-10','10-20','20-40','40-60','60-90']

player_active_days_30['Active_Cohort'] = pd.cut(player_active_days_30['ActivePlayerDays'], bins=bins, labels=labels, include_lowest=True)
```

```
player_active_days_30
```

**Which cohort contributes the most to total deposits, and what could explain that distribution?**

```
First_Deposit.head()
```

```
merged_df = pd.merge(player_active_days_30, First_Deposit,
                     how='left',
                     on='Src_Player_Id')
```

```
merged_df.head()
```

```
cohort_contribution = merged_df.groupby('Active_Cohort')['First_Deposit_Amount'].sum().reset_index().sort_values(by='First_Deposit_Amount', ascending=False)
```

```
cohort_contribution.head()
```

**Observation:**

- Cohorts (1-10 days) often contribute more to total deposits.

- Reason: More active players are likely engaged, increasing probability of depositing higher amounts.

### **Distribution of First Deposit Amounts**

```
bins = [0,100,1000,13000]
labels = ['Low','Medium','High']

total_deposit['Deposit_Bin'] = pd.cut(total_deposit['First_Deposit_Amount'], bins=bins, labels=labels, include_lowest=True)


deposit_distribution = total_deposit['Deposit_Bin'].value_counts().sort_index()




plt.figure(figsize=(8,5))
ax = deposit_distribution.plot(kind='bar', color='skyblue')

for i, v in enumerate(deposit_distribution):
    ax.text(i, v + 50, str(v), ha='center', fontweight='bold')  # adjust +50 as offset

plt.xlabel('First Deposit Amount (Bins)')
plt.ylabel('Number of Players')
plt.title('Distribution of First Deposit Amounts')
plt.show()

```

**First Deposit Amount Distribution**

We binned players first deposit amounts into three categories:

- Low: 0  100

- Medium: 101  1000

- High: 1001  13,000

The bar chart shows the number of players in each bin, with counts labeled on top of each bar for clarity.

```
total_deposit.First_Deposit_Amount.value_counts()
```

**Observations (Clustering)**

- There is a noticeable clustering around the Low and Medium bins.

- This clustering often occurs in datasets because:

  - Players tend to deposit round numbers like **( 25, 30, 50, 80, 100, 500 etc)**.

  - Minimum deposit limits or promotional thresholds set by the platform encourage specific amounts.

  - Marketing campaigns or bonuses often target specific deposit tiers.
