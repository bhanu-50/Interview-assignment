# **Data correction**

```
#Removing the unwanted columns
tables = {'BonusCost':BonusCost,'First_Bet':First_Bet,'First_Deposit':First_Deposit,'Player_Activity':Player_Activity,'Player_Details':Player_Details}
for i in tables.values():
    if i['Unnamed: 0'].isna().sum()==len(i):
       i.drop('Unnamed: 0', axis=1, inplace=True)
```

```
#Removing the extra metadata
for i in tables.values():
  i.columns = i.iloc[0]
  i.drop(0,inplace=True,axis=0)
```

```
#Basic info of datasets
for name,i in tables.items():
  print("--"*30)
  print(f'{name}:',end="\n\n")
  print(i.info())
  print("--"*30)
```

# **Data Type Cleanup Before Analysis**

```
# Converting columns to specific types for further usage
BonusCost = BonusCost.astype({
    "Src_PLAYER_ID": "int64",
    "ActivityMonth": "string",
    "Product": "string",
    "BONUS_COST": "float"
})

First_Bet = First_Bet.astype({
    "Src_Player_Id": "int64",
    "System_First_Bet_Datetime": "string",
    "System_First_BetSlip_Amt": "float",
    "System_First_Bet_Product_Group": "string",
    "System_First_Bet_Product": "string",
    "System_First_Bet_Channel": "string",
    "System_First_Bet_Platform": "string"
})

First_Deposit = First_Deposit.astype({
    "Src_Player_Id": "int64",
    "First_Deposit_Date": "string",
    "First_Deposit_Channel": "string",
    "First_Deposit_Method": "string",
    "First_Deposit_Amount": "float"
})


Player_Activity = Player_Activity.astype({
    "src_player_id": "int64",
    "ActivityMonth": "string",
    "Product": "string",
    "SubProduct": "string",
    "ActivePlayerDays": "float",
    "Bet_Amount": "float",
    "Win_Amount": "float",
    "Gross_Win": "float",
    "Net_Gross_Win": "float"
})

Player_Details = Player_Details.astype({
    "Src_Player_Id": "int64",
    "Signup_Date": "string",
    "acquisition_channel": "string",
    "src_advertiser_id": "string",
    "Gender": "string",
    "Date_Of_Birth": "string",
    "Internal_Player_YN": "string"
})
```
